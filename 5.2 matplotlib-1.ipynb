### <font color=blue>3. Matplotlib的绘图体系</font>
- Matplotlib的图像是画在 <font color=red>*figure*</font> 上的，每一个figure又包含了一个或多个 *axes* 。
    - figure：画布。可以创建多张画布，一般正常状况下一张画布占据一个图形窗口。
    - axes：坐标域：绘图区/子图/图形区/坐标区，一个可以指定坐标系的子区域集合。
            存在于画布中。即由坐标轴划定的图形区。一张画布中可以有多个绘图区（子图）。
    - title、xlabel、ylabel、legend等：绘图区标题、轴标签等存在于标签区。
            由axes对象管理，它们的作用是对某一绘图区进行说明。

#### ⚡⚡<font color=blue>Matplotlib 有两种绘图方式</font>
- **pyplot-style ：** `plt.xxx系列`，是通过matplotlib提供的api绘图，Pyplot提供了很多[基本的function](#id_plt_functions)可以让你很快地画出图来。
            - 当第一次执行plt.xxx()绘图语句时，系统会判断是否已经有了figure对象；
                * 如果没有，自动隐式创建一个 figure，及其一个 axes，即此时我们只能绘制一个图形。  
                * 如已显式创建figure，当有多个axes时，记住一个原则即可：plt 的操作默认是针对当前的 figure 和 axes。
            - 但是如果你想要更细致的精调图表细节，就要使用另外一种方法。
- **OO-style ：** `fig, ax = plt.subplots()`，面向对象范式(OO styple)的绘图方式，[指定 figure和 axes，然后对axes单独操作](#id_use_fig+axes)。
            通过显式创建figure(控制总体图形大小)，axes(控制绘图，坐标等), line, text等对象，然后调用这些对象的方法来进行精细的控制和调整。 
---

隐式创建和显示创建  
* 优势：如果只是绘制一个小图形，那么直接使用plt.xxx()方式，会自动帮我们创建一个figure对象和一个axes坐标系，图形最终就是绘制在这个axes坐标系之上。
* 劣势：如果想要在一个figure对象上，绘制多个图形，那么必须能够拿到并调用每个子图上的axes对象，然后[分别在子图对应位置的坐标系上进行绘图](#id_drawing_hierarchy)。
        注意：如果figure对象是被默认创建的，那么根本拿不到axes对象。因此，需要显示创建figure对象。

<a id="id_create_subplot"></a>  

2. **创建子图，布局绘图区**   &emsp;[📂步骤](#id_drawing_steps)  
    - 在使用 Matplotlib 绘图时，可以将一张画布划分为若干个子区域，然后我们就可以在这些区域上绘制不同的图形。
    - ---
    - figure对象提供 [add_subplot()](#id_addsubplot) 函数，为 figure 添加子图，不覆盖之前的图。  
        函数格式：<code><mark><font color=blue size=4>ax = figure.add_subplot(nrows, ncols, index)</font></mark></code>  
           将整个绘图区域划分成几行几列（nrows*ncols）的子区域，并在 index 指定的位置创建坐标区。  
           按照从左到右，从上到下的顺序对每个子区域进行编号(1...N)，左上的子区域编号为 1、右下的区域编号为 N；编号可以通过参数index来设置，用来选定具体的某个子区域。
        
    - 使用 pyplot模块提供的 subplot() 和 subplots() 方法来绘制多个子图。
        - [subplots()](#id_subplots) 函数，它创建了一个包含子图区域的画布，同时创建子图坐标区。  
            函数格式：<code><mark><font color=blue size=4>fig , ax = pyplot.subplots(nrows, ncols)</font></mark></code> 
                函数的返回值是一个元组，包括一个画布对象和一个axes对象数组。其中 axes对象的数量等于 nrows * ncols，且每个axes对象均可通过索引值访问。
        - [subplot()](#id_plot-subplot) 函数，它可以均等地划分画布，在绘图时需要指定位置，切换为当前子图。  
            函数格式：<code><mark><font color=blue size=4>ax = pyplot.subplot(nrows, ncols, index)</font></mark></code>  
                将整个绘图区域分成 nrows 行和 ncols 列，而index用于对子图进行编号。函数返回一个表示当前子图的Axes对象，可以用来对子图进行各种设置和绘图操作。

```python
# 用subplot() 方法来绘制多个子图。
'''
## 没有使用plt.subplots(1,2)创建子图布局
#  创建了一个1x2的subplot对象，即两块绘制区域，
#     当我们在每个区域进行绘图时，需要使用第三个参数index来指定绘制子区域。
# 当有多个axes时，记住一个原则即可：plt 的操作默认是针对当前的 figure 和 axes 。
'''
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
plt.subplot(1, 2, 1)
plt.plot(x1, y1, 'ro-')  # "ro-"中r表示红色，o表示圆圈，-表示实线; 【可以写成 o-r 或 or- 或 -or 顺序不要求】
plt.title("plot 1")

#plot 2:
x2 = np.linspace(0.0, 2.0)
y2 = np.cos(2 * np.pi * x2)
plt.subplot(1, 2, 2)
plt.plot(x2, y2, '.-')  # ".-"中 .表示点标记，-表示实线，
plt.title("plot 2")

plt.suptitle("subplot Test")  #设置图表总标题（super title），位于整个图表的最顶部，通常用于描述整个图表的主题

plt.show()
```

**Matplotlib绘图的 Figure 和 Axes - 概念总结** 
1. Figure ： 
  fig = plt.figure()
    - 画图的第一件事，就是创建一个画布figure，然后在这个画布上加各种元素。
2. Axes ：
  ax = fig.add_subplot(1,1,1)
    - 指定了一个有数值范围限制的绘图区域。
    - 在一个给定的画布（figure）中可以包含多个 axes 对象，但是一个 axes 对象只能在一个画布中使用。

        - 首先，这个不是你画图的xy坐标轴！  
        - 如果你的figure只有一张图，那么你只有一个axes。  
          如果你的figure有subplot，那么每一个subplot就是一个axes。  
              axes是matplotlib的宇宙中心！在axes下可以修改编辑的变量非常多，基本上能包含你的所有需求。  
        - Axis (ax.xaxis / ax.yaxis)：  这才是你的xy坐标轴。
            每个坐标轴（axis）是由刻度线（tick）和刻度标签（label）组成的。
                每一个坐标轴实际上也是由一系列的子图（subplots）组成的，每一个子图表示了一个刻度线和对应的刻度标签。
                这些子图实际上也是axes对象，可以通过ax.xaxis获取到表示x轴的axes对象。然后可以对其进行编辑，更改样式、调整刻度线、标签等属性，就会修改xaxis在图形上的表现。
