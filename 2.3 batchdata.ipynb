## 2.3.1 字符串
#### 字符串：字符类型的数据序列
#### 字符串的常见操作：
1. 字符串拼接: +
2. 字符串截取: 切片
3. 字符串查找: 遍历/find

#### 1. 字符串拼接
    使用“+”运算符可以连接多个字符串并产生一个新的字符串对象。

```python
#   【例2-5】

Str1 = '我今天敲了'
Num1 = 10000
Str2 = '行代码，我想换行……'
PG1 = '程序员甲：'+ Str1 + str(Num1) + Str2    #拼接出第一个字符串
PG2 = '程序员乙：敲一下回车键啊'               #第二个字符串
print(PG1 + '\n' + PG2)                        #以回车分隔，显示两个字符串
```

#### 2. 截取字符串

   - 切片的语法格式如下：
        <font color=green>**string[start:end:step]**</font>
   - 参数说明：
    - <font color=blue>string:待截取的字符串。</font>
    - <font color=blue>start:要截取的第一个字符的索引，默认为0。</font>
    - <font color=blue>end: 要截取的最后一个字符的索引，默认为字符串的长度。</font>
    - <font color=blue>step:切片操作的步长，默认为1。</font>
    




```python
#   字符串索引和切片
word = '南京市长江大桥'
ans1 = word[0:3]  #截取前三个字符
print(ans1)
print(word[-4:-1] + word[-1])  #倒数第二、三、四个字符，然后拼接最后一个字符
```

```python
#   【例2-6】 2521090307
'''红星商场的商品编码由不同含义的10位字符组成，输入一个符合标准的10位商品编码，通过程序自动识别出商品产地和出厂日期。
其中，商品编码各个位置字符的含义如下：
第1位：值为0表示公司自产，值为1表示国产，值为2表示进口。
第2位：商品的类别
第3~8位：商品的出厂日期
第9~10位：表示商品的编号。'''

while True:    #持续判断，直到运行break语句
    mystr=input("请输入一个10位产品编码（按quit退出）")  #获取用户输入
    if(mystr=="quit"):    #判断是否退出
        break
    else:
        if mystr[0]=='0':  #根据第一位判断产地
            print("公司自产")
        elif mystr[0]=='1':
            print("国产")
        else:
            print("国外进口")
    pyear=mystr[2:4]+"年"  #第三、四位为年份
    pmonth=mystr[4:6]+"月" #第五、六位为月份
    pday=mystr[6:8]+"日"   #第七、八位为日期
    print("商品的生产日期是"+pyear+pmonth+pday)
```

#### 3. 字符串检索
    在字符串中查找特定的字符，既可以通过对字符串进行遍历，又可以通过字符串的函数进行处理。

```python
#   【例2-7】
'''在车牌数据的集合中，根据输入的车牌起始字符，找到完整的车牌号码。'''

carNoSet="沪A39201；京C39；津AT39192；皖A6C36；渝79157；冀2044J；晋ZE875；蒙201075；辽0S050；吉A9K979；黑E0H009；苏F9K979；浙J30756；"
str1=input('请输入车牌第一个字符（如：沪、京、津、皖、渝、冀、晋、蒙、辽、吉、黑、苏、浙）：')   #获取用户输入
i=carNoSet.find(str1) #在车票数据中查找用户输入的字符，并返回所在位置
if (i<0):   #如果没有找到该字符
    print('未找到')   
else:
    j=carNoSet[i:].find("；")  #自该字符位置开始，找第一个“;”分号
    carNo=carNoSet[i:i+j]    #读取字符位置到最近的分号之间的字符
    print(carNo)
```

## 2.3.2 列表
    Python中的列表可以包含整数、实数、字符串甚至列表等不同类型的数据，列表是非常灵活的数据类型。

1. 遍历访问一维列表
    - 一维列表可以存储一维、线性数据。在实际应用中，经常需要对列表中的数据项进行遍历（也称为迭代）。
    - 常用的列表迭代方法有三种：
      - for 循环遍历
      - 通过迭代器遍历：enumerate(iterable, start=0）</br>
           函数返回一个迭代器，能为可迭代对象(如列表、元组或字符串)生成一个索引序列，同时列出数据和数据下标。
      - 按下标遍历
      
2. 遍历访问二维列表
    - 二维列表可以存储二维、表格数据。
    - 在访问二维列表中的元素时，需要指定第一和第二列表的索引。二维列表的存取需要使用行和列两个下标。
    


```python
#   【例2-8】列表的遍历

lis = ['休斯顿火箭','金州勇士','波特兰开拓者','犹他爵士','圣安东尼奥马刺']

print('-----(1)直接遍历-------')
for item in lis:
    print(item)
  
print('-----(2)使用enumerate函数遍历-------')
for i in enumerate(lis):
    print(i)

#对于列表类型，还有一种通过下标遍历的方式，如使用range()函数
print('-----(3)按下标遍历-------')
for i in range(len(lis)):
    print(lis[i])
```

## 2.3.3 元组
    元组写在小括号( )里，元素之间用逗号分隔，元素可以具有不同的类型。
    元组（Tuple）与列表类似，但元组的元素不能修改。
    
- 元组的访问与列表、字符串类似，元组和列表、字符串对比，也存在区别。
  - （1）元组比列表操作速度快，遍历操作效率更高。
  - （2）元组不可修改，能够对不可修改的数据进行保护，使数据更加安全。
  - （3）元组可以用作字典中的key，但列表不行。

```python
#   【例2-11】元组的访问

tuple = ( 'SpiderMan',2017 ,33.4, 'Homecoming', 14 ) 
tinytuple = (16, 'Marvel') 
print (tuple)      # 输出完整元组 
print (tuple[0])    # 输出元组的第一个元素 
print (tuple[3:4])   # 输出从第二个元素开始到第三个元素 
print (tuple + tinytuple) # 连接元组
```

```python
#   【例2-12】修改元组中的List类型数据项
''' 虽然元组的元素不可改变，但如果元组内部的数据项是可变的类型，则该数据项可以修改。'''

tuple = ([16, 'Marvel'] , 'SpiderMan',2017 ,33.4, 'Homecoming', 14,) 
print(tuple[0])
tuple[0][0]='Marvel'
tuple[0][1]='16'
print (tuple)
```

## 2.3.4 字典

- 字典是一种可变容器类型，且可存储任意类型对象。以<font color=red>键值对（Key-Value Pair）</font>存储数据，并且可以根据键来快速检索值。
- 字典使用大括号 { } 定义, 格式如下所示：<br>
    - dt = { key1 ：value1, key2 ：value2 }<br>
    - 字典的每个键和值 (key/value)之间用冒号分隔，键值对之间用逗号分隔。<br>
    - <font color=red>键是唯一的</font>，如果出现了重复，则后面的键值对会替换前面的键值对。值的数据及类型不限，可以是字符串、数字或元组。
- 字典对象的常用方法



#### 1.字典的访问 
      访问字典中的值需要使用字典的键值，格式为：dict[key]


```python
''' 1.字典的访问 
      访问字典中的值需要使用字典的键值，这个键值用方括号括起来，格式为：
      dt[ key ]
'''
#  【例2- 】字典的访问
dict1 = {'Name': 'Mary', 'Age': 7, 'Class': 'First'}; 
print(dict1);
print("Name: ", dict1['Name'])
print("Age: ", dict1['Age'])
```

```python
#   【例2-13】列表可以作为字典的value值

dict1 = {'Name': ['Mary','Tom','Philp'], 'Age': [7,8,9], 'Class': ['1st','2nd','3rd']}; 
print(dict1);
print("Name: ", dict1['Name'])

dict1['Age'] = [8,9,10]
print("Age: ", dict1['Age'])
```

```python
#   【例2-14】元组可以作为字典的key值

dict1={('Mon','Thu'):'语文',('Tue','Wed','Fri'):'数学'}                                     
dict1[('Mon','Thu')]
```

#### 2. 修改字典
    可以向字典添加、修改或删除键/值对。


```python
#   【例2-315】用update()函数修改字典。

dict1 = {'Name': 'Zara', 'Class': 'First'}

#添加add
dict1['Gender'] = "Female"
print(dict1)

#修改update
dict1.update({"No":"001"})
print(dict1)

#也可以使用update方法添加/修改多个数据
dict1.update({'Gender':"F","Id":1})
print(dict1)
```

#### 3. 删除字典元素
    删除一个字典键值对用del命令, 清空字典用clear命令。


```python
#   【例2-16】删除字典元素。

# 删除一个元素
del dict1['Gender']
print(dict1)

# 清空字典
dict1.clear()
print(dict1)
```

#### 4. 字典数据的遍历


```python
#   【例2-17】遍历字典中的数据。

dictP = {'朱自清':['《春》','《匆匆》','《绿》','《背影》','《荷塘月色》'],
        '老舍':['《济南的冬天》','《茶馆》','《骆驼祥子》'],
        "李白":['《静夜思》','《赠汪伦》','《望天门山》','《蜀道难》']
                }

key = input("请输入作家姓名：")   #获取用户输入

if key in dictP.keys():     #如果找到对应的键(key)
    for item in dictP[key]: #读取该键对应的值(value)
        print(item)
else:
    print ("没有找到要查询的作家！")
```

```python
#   【例2-18】访问字典的键、值和键值数据。

dict1={('Mon','Thu'):'语文',('Tue','Wed','Fri'):'数学'}

print('dict1.keys:',dict1.keys())
print('dict1.items:',dict1.items())
print('dict1.values:',dict1.values())

print()
print('---遍历键----------------')
for key in dict1.keys():
    print('key:',key)
    #print('value:',dict1[key])
    
print('---遍历值----------------')
for value in dict1.values():
    print('value:',value)
```

## 2.3.5 集合
- Set(集合)由一列 <font color=red>无序的、不重复</font> 的数据项组成。与数学中的集合概念相同，集合中每个元素都是唯一的。
- 同时，集合不设置顺序，每次输出时元素的排序可能都不相同。
- 集合使用大括号，形式上和字典类似，但数据项不是成对的。

#### 1. Set集合的遍历
    集合中的元素可遍历进行访问，可以使用直接遍历，也可以使用enumerate迭代器进行遍历。不过，集合类型不支持 range()方式的遍历。


```python
#   【例2-19】集合anml为{'紫貂','松貂','青鼬','狼獾’}，对anml集合进行遍历。

#方法一：
anml ={'紫貂','松貂','青鼬','狼獾'}
for item in anml:
       print(item)
        
#方法二：
anml ={'紫貂','松貂','青鼬','狼獾'}
for item in enumerate(anml):
       print(item)
```

#### 2. Python集合操作
    数学集合运算符与Python集合操作符对比表
 

```python
#   【例2-20】集合的交集、并集（合集）、差集。

#分别构造獾和貂两个集合
Huan = {'猪獾','蜜獾','狼獾',}
Diao = {'紫貂','松貂','美洲水鼬','狼獾'}

#交集
DiaoXiong = Huan & Diao
print('貂熊是：',DiaoXiong)

#并集
Youke = Huan | Diao
print('鼬科的是：',Youke)

#差集
DiaoT = Diao - Huan
print('除去獾的貂类：',DiaoT)
```
