## 3.2.1 函数
### 1. 什么是函数  
  - 一段有特定功能、可以<u><font color=red>重复使用</font></u>的代码  
  - 以固定的格式<u><font color=red>封装</font></u>（包装）成一个独立的代码块 
  
### 2. 内置函数 和 自定义函数  
  - 内置函数：前面章节中我们已经接触过多个函数，比如 input() 、print()、range()、len() 函数等等，可以直接使用。  
  - 自定义函数：用户根据自己想要功能创建的函数。 
  
### 3. 函数的定义  
   *语法格式：*<code><font color=red size=5>
    **def** 函数名(<font color=green size=5>参数列表</font>):  </font><font color=blue size=4>
       '''
       为函数提供说明文档
       '''
       //实现特定功能的多行代码
       **函数体**
       [**return** [返回值]]</font> </code>  
   
- <u>函数代码块以 def 关键词开头</u>，后接函数标识符名称和圆括号()。
- <u>形参列表：设置该函数可以接收多少个参数</u>，多个参数之间用逗号 , 分隔。
- <u>函数内容以冒号起始，并且缩进</u>。
- 函数的第一行语句可以选择性地使用文档字符串，用于存放函数说明。
- [return [表达式]]: <u>结束函数，选择性地返回一个值给调用方</u>。不带表达式的return相当于返回 None。  
      其中，用 [] 括起来的为可选择部分，即可以使用，也可以省略。
 


```python
#   【例3-13】创建一个名为Hello的函数，其作用为输出“欢迎进入Python世界”的字符内容。
def  Hello():
    print("欢迎进入Python世界 ")
```

```python
#   【例3-14】定义一个名为Sum()的函数，其作用是计算n 以内的整数之和（包含n）。
def SumN(n):
    s = 0
    for i in range(1, n+1):
        s = s+i
    return s
```

### 5. python 函数的参数传递
####  <font color=red>形参与实参</font>
- 函数参数在定义函数时放在函数名称后面的一对小括号中，在使用函数时，经常会用到<u>形式参数</u>（形参）和<u>实际参数</u>（实参），两者都称为参数。
- 形式参数和实际参数在作用上的区别如下：<br>
  ❖ <font color=blue>形式参数</font>：在定义函数时，函数名后面括号中的参数为“形式参数”，也称形参。<br>
  ❖ <font color=blue>实际参数</font>：在调用一个函数时，函数名后面括号中的参数为“实际参数”。也就是将函数的调用方传递给函数的参数称为实际参数，也称实参。

```python
#   【例3-16】判断用户输入的手机号码是否合法。

def phone(num): #定义phone函数进行判断
    if num.isdigit():
        if len(num)==11:
            print('输入的手机号码合法!')
        else:
            print('输入的手机号码位数有误!')
    else:
        print('入的手机号码含有非法字符!')


instr = input('请输入11位手机号码:')
phone(instr) #调用phone函数完成判断
```

####  值传递和地址传递
<font color='blue'>*实参是如何传递给形参的？*</font>  
1. 调用时的<u>参数个数必须和声明时的一样</u>。  
2. Python 中，根据实际参数的类型不同，函数参数的传递方式可分为 ：值传递和引用（地址）传递。
  - <font color='blue'>值传递</font>：适用于实参类型为不可变类型（字符串、数字、元组）；
  - <font color='blue'>引用（地址）传递</font>：适用于实参类型为可变类型（列表，字典）；

```python
def update2(obj) :
    obj += obj
    print("\t形参值为：", obj)
    
print("------值传递-----")
a = 10
print("a的值为：", a)
print("a作为参数调用函数～")
update2(a)
print("\ta的值(实参值)为：", a)

print("-----引用传递-----")
a = [1,2,3]
print("a的值为：", a)
print("a作为参数调用函数～")
update2(a)
print("\ta的值(实参值)为：", a)
```

## 3.2.2 模块与python库

### 1. 模块  
- **模块(Module)，是一个 Python 文件，以 .py 结尾，包含了 Python 对象定义和Python语句。**  
  模块就是 Python 程序，即任何 Python 程序都可以作为模块 
-
- **从文件级别有逻辑地组织程序，提高代码的可维护性，提高代码的可重用性。**  
<font color=blue size=2>代码的可重用性体现在，当编写好一个模块后，只要编程过程中需要用到该模块中的某个功能（由变量、函数、类实现），无需做重复性的编写工作，直接在程序中导入该模块即可使用该功能。</font>
- 
- 👉👉👉模块，可以理解为是对代码更高级的<font color=red>封装</font>，即把能够实现某一特定功能的代码编写在同一个 .py 文件中，并将其作为一个独立的模块，这样既可以方便其它程序或脚本<font color=red>导入并使用</font>，同时还能有效<font color=red>避免函数名和变量名发生冲突</font>。


### 2. 内置模块和非内置模块  
#### （1）内置模块  
   - <font color=blue> *启动Python时系统会自动导入*</font>，任何程序都可以直接使用。  
   - 内置函数是通过_builtins_模块提供。  <br>
       print( )、sum( )、str( )、enumerate( )...
   
#### （2）非内置模块    
  - 提供功能的扩充  
  - 需要手动导入  


#### （3）模块的导入  
   ######  导入模块的语法：
1. <font color=red size=4>import 模块名1 [as 别名1], 模块名2 [as 别名2]，… </font><br>
  - 导入指定模块中的所有成员（包括变量、函数、类等）。
  - 当需要使用模块中的成员时，需用该模块名（或别名）作为前缀，否则 Python 解释器会报错。<br>
    eg: <code>模块名.函数名 <br>    import time <br>    time.sleep(1)</code>
2. <font color=red size=4>from 模块名 import 成员名1 [as 别名1]，成员名2 [as 别名2]，… </font>
  - 只导入模块中指定的成员，而不是全部成员，到当前命名空间。
  - 当程序中使用该成员时，无需附加任何前缀，直接使用成员名（或别名）即可。<br>
    eg: <code>from time import sleep<br>    sleep(1)</code>

```python
#   【例3-17】导入math整个模块
import math

#使用模块名作为前缀来访问模块中的成员
print(math.pi)   #使用math模块的pi常数显示π
print(math.pow(math.pi,2)) #使用pow()函数计算π的平方
print(math.sqrt(16))  #使用sqrt()函数计算16的平方根
```

```python
#   【例3-18】使用from方法只导入math模块的一部分
from math import sqrt, pi

print(pi)
print(pow(pi,2))
print(sqrt(16)) 
```

```python
#   【例3-21】简易微信群发红包模拟程序。

'''任务：在微信群中有6个好友，分别是小明、李梅、小龙、冬冬、小胖、小巧。小明发出一个指定金额的红包后，6个人可以获得随机的红包。
输出每个人的红包、和手气最佳者。

常用的格式化字符有四种
%s    -- 字符串类型
%d    -- 整数类型(%05d,指定整数的显示位,不足的位数用0补齐)
%f    -- 浮点型(%.2f,指定小数点后显示两位)
%%    -- 输出%
'''
import random

friends = ['小明','李梅','小龙','冬冬','小胖','小巧']
xm = '小明'
```

```python
#1.输入红包总金额
sum_money = float(input('请输入红包金额（1-200）：'))
send_n = 6  # 发红包个数
print('%s发了一个总金额%d元的红包，共分成%d份' % (xm, sum_money, send_n))
max_hongbao = 0    #手气最佳者的金额
recieve_max = ''   #全场手气最佳者
```

```python
#2.划分红包，使总和为红包金额
listTemp = []  #存放6个是随机数
HBlist = []    #存放6个是红包金额
sumHB = 0
for i in range(0,send_n):    #执行6次
    x0 = round(random.uniform(1,100),2)  #得到6个随机数作为红包金额
    listTemp.append(x0)
    sumHB = sumHB+x0      #6个是随机数之和
ratio = sum_money/sumHB   #红包总金额/随机数之和
# print(sumHB, listTemp)
# print(sum_money, HBlist)
```

```python
#3. 用比例计算实际随机红包值
for i in range(0,send_n):
    HB = round(ratio*listTemp[i],2)     #随机数做同比缩放，使6个数之和等于红包总金额
    HBlist.append(HB)                 #保存红包
# print(ratio)
# print(sumHB, listTemp)
# print(sum(HBlist), HBlist)
```

```python
#4.发放红包
for i in range(0, send_n):    
    name = friends[i]    
    if i == send_n-1:  #最后一个红包  
        recieve_m = sum_money     # 剩下的发到最后一个红包    
    else:        
        recieve_m = HBlist[i]     # 在列表中抽取对应红包        
    sum_money = sum_money - recieve_m   # 每发一次红包后剩下的钱 
    print('%s抢到了%.2f元钱'%(name, recieve_m))
    
    if i < send_n-1:               # 显示未被抢红包数量
        print('还有%d个红包没有被抢(剩余%.2f)' %(send_n-1-i, sum_money))
        print() 
    else:        
        print('最后一个红包被%s抢走了!'%name)  # 提示红包已经被抢光了
        
    if recieve_m > max_hongbao:   #记录每一轮最大红包
        max_hongbao = recieve_m
        recieve_max = name      
        
print('%s全场手气最佳,抢得最大红包%.2f元'%(recieve_max,max_hongbao))
```
