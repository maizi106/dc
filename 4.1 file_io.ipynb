**1. 文件路径** <br>
&emsp;文件的两个关键属性：

&emsp;文件的绝对路径与相对路径：
   - <font color=blue >*当前工作目录*</font>   
    每个运行在计算机上的程序，都有一个“当前工作目录”。所有没有从根文件夹开始的文件名或路径，都假定在当前工作目录下。
   - <font color=blue >*绝对路径*</font>：完整路径名总是从根文件夹开始。例如：<u>*c:\data\cust.txt*</u>。    
    <font color=blue >*相对路径*</font>：指的是文件相对于当前工作目录所在的位置。例如当前路径为 *c:\data*，文件 *c:\data\ex1\calu.txt* 可以表达成 <u>*\ex1\calu.txt*</u> 形式。

### 4.1.2 读取文件内容 [◀](#file_operate)
读取文件数据的前提是，使用 open() 函数指定[打开文件的模式](#file_open_mode)必须为可读模式（包括 r、rb、r+、rb+ 4 种），默认是文本读取(rt)模式。  
Python 提供了如下 3 种函数，它们都可以实现读取文件中数据的操作：  
- <code><font color=blue size=4>read([size])</font></code> 函数：逐个字节或者字符读取文件中的内容；返回值为字符串。
              可选参数size，用于指定一次最多可读取的字符（字节）个数；默认为 -1，表示读取整个文件。
- <code><font color=blue size=4>readline([size])</font></code> 函数：逐行读取文件中的内容；返回值为字符串。读取文件中的一行时，包含最后的换行符“\n”。
- <code><font color=blue size=4>readlines( )</font></code> 函数：一次性读取文件中所有内容，返回值为列表，其中每个元素为文件中的一行内容。

 
1. <font color=blue>**※**</font>  调用read()会一次性读取文件的全部内容，  
        如果文件很大，内存就爆了，所以可以反复调用read(size)方法，每次最多读取size个字节的内容。  
        如果文件很小，read()一次性读取最方便；  
        如果不能确定文件大小，反复调用read(size)比较保险；如果是配置文件，调用readlines()最方便。
2. <font color=blue>**※※ [读取文件时数据经常含有特殊字符](#op_sepcial_char)，\n（换行）、\r（回车)、\t（制表）、'&emsp;'（空格）**</font>等。   

**read()，读取文件内容**

```python

# 【例4-3】read()函数读取整个文件。

with open("data.txt") as f:   # 默认模式为‘r’，只读模式
    ct10 = f.read(5)          # 读5个字符
    print(ct10)
    print('======')
    contents = f.read()       # 从当前位置，读文件全部内容
    print(contents)  

```

**readline()，逐行读取文件中的内容，返回一行**

```python
# 【例4-4】使用readline()函数逐行读取。

with open('data.txt') as f:     
    line1 = f.readline()  # 读取第一行数据（此时已经指向第一行末尾）
    line2 = f.readline()  # 从上一次读取末尾开始读取（第二行）
    
    print('------line1------')
    print(line1)
#     print(line1.strip())
#     print(line1.split())
    
    print('------line2------')
    print(line2)
#     print(line2.strip())

print('-----------------')
    
```

**readlines()，一次性读取文件中所有内容，逐行返回为列表值**

```python
# 【例4-5】使用readlines()一次读取多行：

with open('data.txt') as f: 
    lines = f.readlines()  #文件数据读到一个列表，每个元素对应一行
print(lines)               #每一行数据都包含了换行符

print('================================') 
for line in lines: 
    print(line.rstrip())   #使用rstrip()删掉末尾的空格
```

**使用 for 语句迭代读取，每次一行**

```python
# 【例4-6】使用for循环逐行读取文件。

#逐行读取
with open('data.txt') as f:
    for lineData in f:
        print(lineData.rstrip()) # 去掉每行末尾的换行符
        
```

向文件中写入数据，需保证使用 open() 函数是以**‘w’**（写入）或者**‘a’**（追加）模式打开文件，如：r+、w、w+、a 或 a+。采用不同的[文件打开模式](#file_open_mode)，会直接影响 write() 函数向文件中写入数据的效果。  
  
Python 中的文件对象提供了：
- <code><font color=blue size=4>write( )</font></code>函数：将字符串（或字节串，仅适用写入二进制文件中）写入文件，返回的是写入的字符长度。

**'r+' 模式和 'w'模式，写入时的区别**

```python
# r+ 更新模式，文件既存，写入时内容覆盖
# 文件已存在，打开写入数据时，原有文件的数据被覆盖

with open('INFO.txt','r+') as f:       # 'r+'表示读取并修改数据
    #print(f.read())  
    f.write("Attention!!! overwrite！") 
    
with open('INFO.txt','r') as f:
    print(f.read())      
```


### 文件指针  

使用 open() 函数打开文件并读取文件中的内容时，总是会从文件的第一个字符（字节）开始读起。

文件中每个数据（以 b 模式打开，每个数据就是一个字节；以普通模式打开，每个数据就是一个字符）  


- 文件指针用于标明文件读写的起始位置。  
-  seek() 函数用于移动文件指针到文件的指定位置。
- 注意，当向文件中写入数据时，如果不是文件的尾部，写入位置的原有数据不会自行向后移动，新写入的数据会将文件中处于该位置的数据直接覆盖掉。
---
*seek() 函数语法格式：* <br> 
&emsp;&emsp;&emsp;&emsp;<code><font color=blue size=4>file.seek(offset[, whence])</font></code>  
- 主要参数：  
        offset——偏移量。从指定位置开始，需要移动的字节数。  
        whence——指定的位置。可选参数，代表Offset的起始点，默认值为 0。值为0代表文件头，1代表当前位置，2代表文件末尾。  
- 例如：  
        seek(0,0)：表示指针回到文件头；  
        seek(0,2)：表示指针到达文件末尾；  
        seek(num,0)：表示指针从文件头开始，移动num个字节。  

```python
# 补充例题-2
# 打开文件
f = open("write_data.txt", "r+")
print("文件名为: ", f.name)

line = f.readline()
line = f.readline()
print("读取的数据为: %s" % (line))

# 重新设置文件读取指针到开头
f.seek(0,0)
line = f.readline()
print("读取的数据为: %s" % (line))

# 重新设置文件读取指针到末尾
f.seek(0,2)
line = f.readline()
print("读取的数据为: %s" % (line))

# 关闭文件
f.close()
```

**编码**  
- 给 [<font size=3>open( )函数</font>](#file_open) 传入encoding参数，将字符串自动转换成指定编码，如使用 utf-8 编码可以适配中文。  
- 默认都是读取文本文件，并且是UTF-8编码的文本文件。  
        f = open('txt.txt', 'r', encoding='utf-8')  
        f.read()

- 遇到有些编码不规范的文件，你可能会遇到UnicodeDecodeError，因为在文本文件中可能夹杂了一些非法编码的字符。<br>遇到这种情况，open()函数还接收一个errors参数，表示如果遇到编码错误后如何处理。最简单的方式是直接忽略：  
        f = open('gbk.txt', 'r', encoding='gbk', errors='ignore')  

```python
# 【例4-9】读取IMDB电影数据集。

filmlist=[]
file=open('score.txt','r',encoding='UTF-8')
print('编号|\t电影名称\t|\t用户年龄\t|\t用户职业\t|\t评分\t|')
while True:    
    line=file.readline().split('\t')
    if line==['']:
        break
    else:
        print(line[0],'|\t',line[5],'|\t',line[2],'|\t',line[3],'|\t',line[7])
        filmlist.append(line)
print('='*100)
film=input('请输入要查看的电影评分：')
print('|\t电影名\t|\t评分\t|')
if film !='':
    for fm in filmlist:
        #print(fm)
        if fm[5].find(film)>=0:
            print('|\t',fm[5],'\t|\t',fm[7],'\t|')            
```
